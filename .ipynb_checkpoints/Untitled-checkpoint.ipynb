```python
%matplotlib notebook

import numpy as np
import matplotlib.pyplot as plt
import time

def pltsin(ax, colors=['b']):
    x = np.linspace(0,1,100)
    if ax.lines:
        for line in ax.lines:
            line.set_xdata(x)
            y = np.random.random(size=(100,1))
            line.set_ydata(y)
    else:
        for color in colors:
            y = np.random.random(size=(100,1))
            ax.plot(x, y, color)
    fig.canvas.draw()

fig, ax = plt.subplots(1,2,figsize=(20,5))
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_xlim(0,1)
ax[0].set_ylim(0,1)
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].set_xlim(0,2)
ax[1].set_ylim(0,2)
for f in range(5):
    pltsin(ax[0], ['b', 'r'])
    pltsin(ax[1], ['b', 'r'])
    time.sleep(1)
```
